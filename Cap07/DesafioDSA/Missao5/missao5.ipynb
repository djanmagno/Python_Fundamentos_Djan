## Análise Geral de Compras

```python
purchase_file.head()
```

```python
# Numero de Itens exclusivos
num_itens = purchase_file['Item_ID'].nunique()
# Preço médio por produto
preco_medio = purchase_file['Valor'].mean()
# Número Total de Compras
total_compras = purchase_file['Login'].count()
# Valor total arrecadado
total_vendas = purchase_file['Valor'].sum()

print(f'São {num_itens} itens exclusivos.\n')
print(f'O preço médio de é R$ {preco_medio:.3}.\n')
print(f'São {total_compras} compras no total.\n')
print(f'Foram arrecadados um total de R$ {total_vendas} em vendas.')
```

## Análise Demográfica

```python
df_demo = purchase_file[['Login','Sexo']].drop_duplicates()
df_genero = df_demo.Sexo.value_counts().to_frame('Total')
df_genero['Porcentagem'] = round(100*df_genero['Total'] / df_genero['Total'].sum(),2)
df_genero
```

## Informações Demográficas Por Gênero

```python
df_demo_gen = purchase_file[['Sexo','Valor','Idade']]
df_demo_gen.head()
```

```python
result = df_demo_gen.groupby(['Sexo']).Valor.agg(["count", "mean", "sum"])
result = result.rename(columns={'count':'Total','mean':'Média','sum':'Valor Total'})
result['Porcentagem'] = 100*result['Total']/result['Total'].sum()
result.round(2)
```

```python
# > 10, 10-14, 15-19, 20-24, 25-29,30-34, 35-39, < 40
#df_demo_gen.loc[(df_demo_gen.Idade < 10),  'Rotulo'] = 'Menos de 10'

def age_buckets(x):
    if x < 10: 
        return 'Menos de 10' 
    elif 10 <= x <= 14: 
        return '10 a 14'
    elif 15 <= x <= 19: 
        return '15 a 19' 
    elif 20 <= x <= 24: 
        return '20 a 24' 
    elif 25 <= x <= 29: 
        return '25 a 29' 
    elif 30 <= x <= 34: 
        return '30 a 34' 
    elif 35 <= x <= 39: 
        return '35 a 39' 
    elif x >=40: 
        return 'Mais de 40' 
    else: 
        return 'other'

df_demo_gen['Label'] = df_demo_gen.Idade.apply(age_buckets)

df_demo_gen.head()
```

```python
result_2 = df_demo_gen.groupby(['Label']).Valor.agg(["count", "mean", "sum"])
result_2 = result_2.rename(columns={'count':'Total','mean':'Valor Médio Unitário','sum':'Valor Total da Compra'})
result_2['Porcentagem'] = 100*result_2['Total']/df_genero['Total'].sum()
result_2.sort_values(by=['Label']).round(2)
```

## Consumidores Mais Populares (Top 5)

```python
df_consu = purchase_file[['Login','Valor']]
df_consu
```

```python
# Implemente aqui sua solução

result_3 = df_consu.groupby(['Login']).Valor.agg(["count", 
                                                  "mean", "sum"]).sort_values(by='count',
                                                                              ascending=False).round(2).head(5)
result_3 = result_3.rename(columns={'count':'Número de Compras',
                                    'mean':'Valor Médio de Compra','sum':'Valor Total de Compra'})
result_3
```

## Itens Mais Populares

```python
df_item = purchase_file[['Item_ID', 'Nome_do_Item', 'Valor']]
df_item.head()
```

```python
result_4 = df_item.groupby(['Nome_do_Item']).Valor.agg(["count", 
                                                  "mean", "sum"]).sort_values(by='count',
                                                                              ascending=False).round(2).head(5)
result_4 = result_4.rename(columns={'count':'Número de Compras',
                                    'mean':'Valor Médio de Compra','sum':'Valor Total de Compra'})
result_4
```

## Itens Mais Lucrativos

```python
# Implemente aqui sua solução
result_5 = df_item.groupby(['Nome_do_Item']).Valor.agg(["count", 
                                                  "mean", "sum"]).sort_values(by='sum',
                                                                              ascending=False).round(2).head(5)
result_5 = result_5.rename(columns={'count':'Número de Compras',
                                    'mean':'Valor Médio de Compra','sum':'Valor Total de Compra'})
result_5
```
