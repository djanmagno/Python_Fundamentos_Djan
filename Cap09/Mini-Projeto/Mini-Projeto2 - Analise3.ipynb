```python
# Imports
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
sns.set(style = "white")
%matplotlib inline
```

```python
# Dataset
clean_data_path = "dataset/autos.csv"
df = pd.read_csv(clean_data_path,encoding = "latin-1")
```

```python
df.columns
```

```python
df.info()
```

## Preço médio do veículo por tipo de combustível e tipo de caixa de câmbio

```python
pm_comb_camb = df[['gearbox', 'fuelType','price']]
k = pm_comb_camb.groupby(['gearbox', 'fuelType'])['price'].mean().reset_index(name='media')

filtro_a = k['gearbox'] == 'Unspecified'
filtro_b = k['gearbox'] == 'automatik'
filtro_c = k['gearbox'] == 'manuell'

a = k.loc[filtro_a, ['media']]
a = list(a['media'])

b = k.loc[filtro_b, ['media']]
b = list(b['media'])

c = k.loc[filtro_c, ['media']]
c = list(c['media'])

labels = k.loc[filtro_a, ['fuelType']]
labels = list(labels['fuelType'])

x = np.arange(len(labels))  # the label locations
w = 0.30  # the width of the bars

fig, ax = plt.subplots(figsize=(18, 9))
rects1 = ax.bar(x - w, a, color = 'blanchedalmond', width = w, label='Unspecified', align='center')
rects2 = ax.bar(x, b, color = 'gainsboro', width = w, label='automatik', align='center')
rects3 = ax.bar(x + w, c, color =  'greenyellow', width = w, label='manuell', align='center')
ax.xaxis.grid(False) # vertical lines
ax.yaxis.grid()

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_title('Preço Médio dos Veículos por tipo de câmbio e tipo de combustível', fontsize = 21)

ax.set_xlabel('\nTipo de Combustível', fontsize = 18)
ax.set_xticks(x)
ax.set_xticklabels(labels, fontsize = 14)

ax.set_ylabel('Preço Médio', fontsize = 18)
plt.yticks(fontsize = 14)

ax.legend(fontsize = 14)
```

```python
# Salvando o plot
fig.savefig("plots/Analise3/fueltype-vehicleType-price.png")
```

## Potência média de um veículo por tipo de veículo e tipo de caixa de câmbio

```python
# Crie um Barplot com a Potência média de um veículo por tipo de veículo e tipo de caixa de câmbio

pot_tipo_cambio = df[['vehicleType', 'gearbox','powerPS']]
l = pot_tipo_cambio.groupby(['vehicleType', 'gearbox'])['powerPS'].mean().reset_index(name='media')

filtro_u = l['gearbox'] == 'Unspecified'
filtro_au = l['gearbox'] == 'automatik'
filtro_ma = l['gearbox'] == 'manuell'

u = l.loc[filtro_u, ['media']]
u = list(u['media'])

au = l.loc[filtro_au, ['media']]
au = list(au['media'])

ma = l.loc[filtro_ma, ['media']]
ma = list(ma['media'])

labels_a = l.loc[filtro_u, ['vehicleType']]
labels_a = list(labels_a['vehicleType'])

x_1 = np.arange(len(labels_a))  # the label locations
w_1 = 0.30  # the width of the bars

fig, ax = plt.subplots(figsize=(18, 9))
rectsa = ax.bar(x_1 - w_1, u, color = 'lightsteelblue', width = w_1, label='Unspecified', align='center')
rectsb = ax.bar(x_1, au, color = 'moccasin', width = w_1, label='automatik', align='center')
rectsc = ax.bar(x_1 + w_1, ma, color =   'silver', width = w_1, label='manuell', align='center')
ax.xaxis.grid(False) # vertical lines
ax.yaxis.grid(False)

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_title('Preço Médio dos Veículos por tipo de câmbio e tipo de combustível', fontsize = 21)

ax.set_xlabel('\nTipo de Combustível', fontsize = 18)
ax.set_xticks(x_1)
ax.set_xticklabels(labels_a, fontsize = 14)

ax.set_ylabel('Potência Média', fontsize = 18)
plt.yticks(fontsize = 14)

ax.legend(fontsize = 14)
```

```python
# Salvando o plot
fig.savefig("plots/Analise3/vehicletype-fueltype-power.png")
```
