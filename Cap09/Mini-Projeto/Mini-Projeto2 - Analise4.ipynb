```python
# Imports
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
sns.set(style = "white")
%matplotlib inline
```

```python
# Dataset
clean_data_path = "dataset/autos.csv"
df = pd.read_csv(clean_data_path,encoding = "latin-1")
```

```python
df.columns
```

```python
# Calcule a média de preço por marca e por veículo

med_pric_veic = df.groupby(['brand','vehicleType'])['price'].mean().\
reset_index(name='media').sort_values(by='brand', ascending=False)

med_pric_veic.head()
```

## Preço médio de um veículo por marca, bem como tipo de veículo

```python
# Para criar o heatmap devemos pivotar a tabela

heatmap_data = pd.pivot_table(med_pric_veic, values='media', index=['brand'],
                              columns='vehicleType').fillna(0).astype(int)

heatmap_data.head()
```

```python
# Crie um Heatmap com Preço médio de um veículo por marca, bem como tipo de veículo

fig = plt.figure(figsize=(13, 26))
sns.heatmap(heatmap_data, cmap="YlGnBu", annot=True, fmt='g')

plt.title("Heatmap - Preço médio de um veículo por marca e tipo de veículo\n", fontsize=21)
plt.xlabel("\nTipo Veículo", fontsize=18)
plt.ylabel("Marca\n", fontsize=18)

plt.tight_layout()
plt.show()
```

```python
# Salvando o plot
fig.savefig("plots/Analise4/heatmap-price-brand-vehicleType.png")
```
